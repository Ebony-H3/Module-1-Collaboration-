```python
#13.1
from datetime import date

today =date.today()

today_string_formatted = today. strftime("%Y-%m-%d")

with open("today.txt","w") as file:
    file.write(today_string_formatted)
print('Date written to today.txt', today_string_formatted)
```

```python
#13.2
with open("today.txt", "r") as file:
    today_string = file.read()

print ("read from file:", today_string)
```

```python
#13.3

from datetime import date
parsed_date = datetime.strptime(today_string.strip(), "%Y-%m-%d")

print("Parsed date:",parsed_date)
print("Type",(parsed_date))
print("Year:", parsed_date.year)
print("Month:",parsed_date.month )
print("Day:", parsed_date.day)
```

```python
import multiprocessing
import os

def whoami(what):
    print("Process %s says: %s" % (os.getpid(), what))

if __name__ == "__main__":
    whoami("I'm the main program")
    for n in range(4):
        p = multiprocessing.Process(target=whoami,
          args=("I'm function %s" % n,))
        p.start()
```
