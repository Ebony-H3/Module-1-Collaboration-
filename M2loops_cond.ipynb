```python
#4.1
secret = 3
while True:
    

    number_guessed = (int(input("Guess a number between 1-10: ")))

    if number_guessed == secret:
        print("Just right!")
        break
    elif number_guessed < 1 or number_guessed > 10:
        print("Please enter a number between 1-10.")
        continue
    elif number_guessed < secret:
        print("Too low!")
    elif number_guessed > secret:
        print("Too high!")
        
print("Thanks for playing!")
```
